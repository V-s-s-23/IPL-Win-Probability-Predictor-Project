```python
def add_1 (x):
    return x+1
```

```python
new_df['total_runs_x'] = new_df['total_runs_x'].apply(add_1)
```

```python
new_df['current_score']=new_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
new_df['runs_left'] = new_df['total_runs_x']-new_df['current_score']
```

```python
new_df['balls_left']= 126 - (new_df['over']*6 + new_df['ball'])
```

```python
new_df['player_dismissed'] = new_df['player_dismissed'].fillna("0")
new_df['player_dismissed'] = new_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
new_df['player_dismissed'] = new_df['player_dismissed'].astype(int)
wicket = new_df.groupby('match_id').cumsum()['player_dismissed'].values
new_df['wicket_left'] = 10-wicket
new_df.tail()
```

```python
# crr = runs/over
```

```python
new_df['crr']=(new_df['current_score']*6)/(120-new_df['balls_left'])
```

```python
# rrr
new_df['rrr']=(new_df['runs_left']*6)/new_df['balls_left']
```

```python
# make result column
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
new_df['result'] = new_df.apply(result,axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
x = final_df.drop(['result'],axis=1)
y = final_df['result']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder = 'passthrough')
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
